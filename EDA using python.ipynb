```python
#potting of HIst graph based on Age Distribution
```

```python
plt.hist(titanic_df['Age'])
plt.xlabel('age')
plt.ylabel('frequency')
plt.title('Age Distribution')
plt.show()
```

```python
#Bar graph plotting of surival counts
survival_count=titanic_df['Survived'].value_counts()
plt.bar(survival_count.index,survival_count.values,color=['green','red'])
plt.title('Survival Counts')
plt.xlabel("Survived")
plt.ylabel("counts")
plt.show()
```

```python
class_count=titanic_df['Pclass'].value_counts()
plt.pie(class_count.values, labels=class_count.index,autopct='%1.1f%%')
plt.title("class distribution")
plt.show()
```
